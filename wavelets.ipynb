```python
import numpy as np
import pywt
import matplotlib.pyplot as plt

# Seal de ejemplo
x = np.linspace(0, 1, 512)
signal = np.sin(2 * np.pi * 5 * x) + np.sin(2 * np.pi * 20 * x)

# Aplicar la DWT (por defecto usa 'db1', Daubechies 1)
coeffs = pywt.wavedec(signal, 'db4', level=4)

# Reconstruir la seal a partir de los coeficientes
reconstructed = pywt.waverec(coeffs, 'db4')

# Grficas
plt.figure(figsize=(10, 6))
plt.subplot(2, 1, 1)
plt.plot(x, signal, label='Seal original')
plt.title("Seal Original")
plt.grid()

plt.subplot(2, 1, 2)
plt.plot(x[:len(reconstructed)], reconstructed, label='Reconstruccin', color='orange')
plt.title("Reconstruccin desde DWT")
plt.grid()

plt.tight_layout()
plt.show()
```
